# WMAZE Model 7
## Models:
### fixed before corr/incorr conditional
### Fixed pairs -- same (AA & CC)
### Fixed pairs -- change (AC & CA)
### "Lost" fixed -- fixed followed by BL trial
### Nonresponse
## Does NOT model:
### Baseline trials
### Conditional trials

```python
import os
from os.path import join
import numpy as np
import pandas as pd
from glob import glob

# Array containing subject ids

#subs = ['WMAZE_001']

subs = ['WMAZE_001', 'WMAZE_002', 'WMAZE_004', 'WMAZE_005', 'WMAZE_006',
        'WMAZE_007', 'WMAZE_008', 'WMAZE_009', 'WMAZE_010', 'WMAZE_012',
        'WMAZE_017', 'WMAZE_018', 'WMAZE_019', 'WMAZE_020', 'WMAZE_021',
        'WMAZE_022', 'WMAZE_023', 'WMAZE_024', 'WMAZE_026', 'WMAZE_027']

# Array containing the three sets
stim_sets = ['set1', 'set2', 'set3']

# Loop to grab correct 6 run text files for each subject
for sub in subs:
    # Array containing path to behavior files
    sub_dir = '/home/data/madlab/data/mri/wmaze/scanner_behav/{0}/'.format(sub)
    # Array containing current sub's 6 behavior file runs
    dir_file = glob(join(sub_dir, '{0}_wmazebl_2015*.txt'.format(sub)))    
    # Sort current sub's txt files in order of run
    dir_file.sort() 
      
    # Loop through each of the set types 
    for i, curr_set in enumerate(stim_sets):
        # Create dataframe for text files to extract EVS
        run1 = pd.read_table(dir_file[i * 2])
        run2 = pd.read_table(dir_file[i * 2 + 1])
        
        run1 = run1[:-3]
        run2 = run2[:-3]
        
        # Convert dataframes into numpy arrays for trial type and correct/incorrect
        # With BLs
        run1_trialtype = run1['TrialType'].values
        run1_correct = run1['Correct'].values
        run1_resp = run1['Resp'].values 
        run2_trialtype = run2['TrialType'].values
        run2_correct = run2['Correct'].values
        run2_resp = run2['Resp'].values
        
        #print run2_trialtype
        

        # Shift the TrialType column up by 1 and insert dummy element into the first index
        run1_trial_shift = run1_trialtype[1:] 
        run1_trial_shift = np.append(run1_trial_shift, -1)
        run2_trial_shift = run2_trialtype[1:]
        run2_trial_shift = np.append(run2_trial_shift, -1)
        #print run2_trial_shift

        # Shift the Correct column up by 1 and insert dummy element into the first index
        # Run 1
        run1_correct_shift = run1_correct[1:]
        run1_correct_shift = np.append(run1_correct_shift, -1)
        # Run 2
        run2_correct_shift = run2_correct[1:]
        run2_correct_shift = np.append(run2_correct_shift, -1)
       
        
        # Run 1
        run1_fixed_before_cond_corr = np.where(((run1_trial_shift == 'B') & (run1_correct_shift == 1)) 
                                               & (run1_trialtype != 'BL'))
        run1_fixed_before_cond_incorr = np.where(((run1_trial_shift == 'B') & (run1_correct_shift == 0)) 
                                                 & (run1_trialtype != 'BL'))
        #two fixeds of the same type
        run1_fixed_2A = np.where((run1_trial_shift == 'A') & (run1_trialtype == 'A') & (run1_resp != 'NR'))
        run1_fixed_2C = np.where((run1_trial_shift == 'C') & (run1_trialtype == 'C') & (run1_resp != 'NR'))
        #two fixed of different types
        run1_fixed_AC = np.where((run1_trial_shift == 'C') & (run1_trialtype == 'A') & (run1_resp != 'NR'))
        run1_fixed_CA = np.where((run1_trial_shift == 'A') & (run1_trialtype == 'C') & (run1_resp != 'NR'))
        #all fixeds followed by a baseline trial
        run1_fixed_lost = np.where(((run1_trialtype == 'A') | (run1_trialtype == 'C')) & (run1_trial_shift == 'BL'))
        run1_nonresponse = np.where((run1_resp == 'NR'))
        #does not model BL trials
        #does not model conditional trials
            
        # Run 2
        run2_fixed_before_cond_corr = np.where(((run2_trial_shift == 'B') & (run2_correct_shift == 1)) 
                                          & (run2_trialtype != 'BL'))
        #print run2_fixed_before_cond_corr
        run2_fixed_before_cond_incorr = np.where(((run2_trial_shift == 'B') & (run2_correct_shift == 0))
                                                 & (run2_trialtype != 'BL'))
         #two fixeds of the same type
        run2_fixed_2A = np.where((run2_trial_shift == 'A') & (run2_trialtype == 'A') & (run2_resp != 'NR'))
        run2_fixed_2C = np.where((run2_trial_shift == 'C') & (run2_trialtype == 'C') & (run2_resp != 'NR'))
        #two fixed of different types
        run2_fixed_AC = np.where((run2_trial_shift == 'C') & (run2_trialtype == 'A') & (run1_resp != 'NR'))
        run2_fixed_CA = np.where((run2_trial_shift == 'A') & (run2_trialtype == 'C') & (run1_resp != 'NR'))
        #all fixeds followed by a baseline trial
        run2_fixed_lost = np.where(((run2_trialtype == 'A') | (run2_trialtype == 'C')) & (run2_trial_shift == 'BL'))
        run2_nonresponse = np.where((run2_resp == 'NR'))
        #does not model BL trials
        #does not model conditional trials
       
        #remove 156 indices if they exist in the Run 1 fixed-fixed arrays                               
        run_types = {'run1_fixed_2A': run1_fixed_2A[0], 'run1_fixed_2C': run1_fixed_2C[0], 
                     'run1_fixed_AC': run1_fixed_AC[0], 'run1_fixed_CA': run1_fixed_CA[0]} 
        
        for key in run_types:
            if len(run_types[key]) > 0:
                if run_types[key][-1] == 156:
                    run_types[key] = run_types[key][:-1]

        
        # All onsets
        run1_onsets = run1['StimOnset']
        run2_onsets = run2['StimOnset']
    

        # Run 1 
        run1_fixed_before_cond_corr_onsets = run1_onsets.values[run1_fixed_before_cond_corr[0]]
        run1_fixed_before_cond_incorr_onsets = run1_onsets.values[run1_fixed_before_cond_incorr[0]]
        #get onsets for separate fixed-same indices
        run1_fixed_2A_onsets = run1_onsets.values[(run_types['run1_fixed_2A'],)]
        run1_fixed_2C_onsets = run1_onsets.values[(run_types['run1_fixed_2C'],)]
        #combine into single sorted array 
        run1_fixed_same_onsets = sorted(np.append(run1_fixed_2A_onsets, run1_fixed_2C_onsets))
        #get onsets for separate fixed-change indices
        run1_fixed_AC_onsets = run1_onsets.values[(run_types['run1_fixed_AC'],)]
        run1_fixed_CA_onsets = run1_onsets.values[(run_types['run1_fixed_CA'],)]
        #combine into single sorted array
        run1_fixed_change_onsets = sorted(np.append(run1_fixed_AC_onsets, run1_fixed_CA_onsets))
        run1_fixed_lost_onsets = run1_onsets.values[run1_fixed_lost[0]]
        run1_nonresponse_onsets = run1_onsets.values[run1_nonresponse[0]]
        
        
        # Run 2 
        run2_fixed_before_cond_corr_onsets = run2_onsets.values[run2_fixed_before_cond_corr[0]]
        run2_fixed_before_cond_incorr_onsets = run2_onsets.values[run2_fixed_before_cond_incorr[0]]
        #get onsets for separate fixed-same indices
        run2_fixed_2A_onsets = run2_onsets.values[run2_fixed_2A[0]]
        run2_fixed_2C_onsets = run2_onsets.values[run2_fixed_2C[0]]
        #combine into single sorted array 
        run2_fixed_same_onsets = sorted(np.append(run2_fixed_2A_onsets, run2_fixed_2C_onsets))
        #get onsets for separate fixed-change indices
        run2_fixed_AC_onsets = run2_onsets.values[run2_fixed_AC[0]]
        run2_fixed_CA_onsets = run2_onsets.values[run2_fixed_CA[0]]
        #combine into single sorted array
        run2_fixed_change_onsets = sorted(np.append(run2_fixed_AC_onsets, run2_fixed_CA_onsets))
        run2_fixed_lost_onsets = run2_onsets.values[run2_fixed_lost[0]]
        run2_nonresponse_onsets = run2_onsets.values[run2_nonresponse[0]]
  
    
        run1_mtrx = np.vstack((run1_onsets,
                               # Numpy array filled with 2.5's
                               np.ones(len(run1_onsets)) * 2.5,
                               # Numpy array filled with 1's
                               np.ones(len(run1_onsets)))).T
      
        run1_fixed_before_cond_corr_mtrx = np.vstack((run1_fixed_before_cond_corr_onsets, 
                                                      np.ones(len(run1_fixed_before_cond_corr_onsets)) * 2.5,
                                                      np.ones(len(run1_fixed_before_cond_corr_onsets)))).T
    
        run1_fixed_before_cond_incorr_mtrx = np.vstack((run1_fixed_before_cond_incorr_onsets, 
                                                        np.ones(len(run1_fixed_before_cond_incorr_onsets)) * 2.5,
                                                        np.ones(len(run1_fixed_before_cond_incorr_onsets)))).T
        
        run1_fixed_same_mtrx = np.vstack((run1_fixed_same_onsets,
                                          np.ones(len(run1_fixed_same_onsets)) * 2.5,
                                          np.ones(len(run1_fixed_same_onsets)))).T
        
        run1_fixed_change_mtrx = np.vstack((run1_fixed_change_onsets,
                                           np.ones(len(run1_fixed_change_onsets)) * 2.5,
                                           np.ones(len(run1_fixed_change_onsets)))).T
        
        run1_fixed_lost_mtrx = np.vstack((run1_fixed_lost_onsets,
                                           np.ones(len(run1_fixed_lost_onsets)) * 2.5,
                                           np.ones(len(run1_fixed_lost_onsets)))).T

        run1_nonresponse_mtrx = np.vstack((run1_nonresponse_onsets,
                                           np.ones(len(run1_nonresponse_onsets)) * 2.5,
                                           np.ones(len(run1_nonresponse_onsets)))).T
        
                
        run2_mtrx = np.vstack((run2_onsets,
                               np.ones(len(run2_onsets)) * 2.5,
                               np.ones(len(run2_onsets)))).T
        
        run2_fixed_before_cond_corr_mtrx = np.vstack((run2_fixed_before_cond_corr_onsets, 
                                                      np.ones(len(run2_fixed_before_cond_corr_onsets)) * 2.5,
                                                      np.ones(len(run2_fixed_before_cond_corr_onsets)))).T
    
        run2_fixed_before_cond_incorr_mtrx = np.vstack((run2_fixed_before_cond_incorr_onsets, 
                                                        np.ones(len(run2_fixed_before_cond_incorr_onsets)) * 2.5,
                                                        np.ones(len(run2_fixed_before_cond_incorr_onsets)))).T
        
        run2_fixed_same_mtrx = np.vstack((run2_fixed_same_onsets,
                                          np.ones(len(run2_fixed_same_onsets)) * 2.5,
                                          np.ones(len(run2_fixed_same_onsets)))).T
        
        run2_fixed_change_mtrx = np.vstack((run2_fixed_change_onsets,
                                            np.ones(len(run2_fixed_change_onsets)) * 2.5,
                                            np.ones(len(run2_fixed_change_onsets)))).T
        
        run2_fixed_lost_mtrx = np.vstack((run2_fixed_lost_onsets,
                                          np.ones(len(run2_fixed_lost_onsets)) * 2.5,
                                          np.ones(len(run2_fixed_lost_onsets)))).T

        run2_nonresponse_mtrx = np.vstack((run2_nonresponse_onsets,
                                           np.ones(len(run2_nonresponse_onsets)) * 2.5,
                                           np.ones(len(run2_nonresponse_onsets)))).T 
        
'''        
        if not os.path.exists(join(sub_dir, 'MRthesis/', 'model7/', 'EVs/')):
            os.makedirs(join(sub_dir, 'MRthesis/', 'model7/', 'EVs/')) 
            
        np.savetxt(sub_dir + '/MRthesis/' + 'model7/' + 'EVs/' + 'run{0}.txt'.format(i * 2 + 1), 
                   run1_mtrx, delimiter = '\t', fmt = '%.4f')
        np.savetxt(sub_dir + '/MRthesis/' + 'model7/' + 'EVs/' + 'run{0}_fixed_before_cond_corr.txt'.format(i * 2 + 1), 
                   run1_fixed_before_cond_corr_mtrx, delimiter = '\t', fmt = '%.4f')      
        np.savetxt(sub_dir + '/MRthesis/' + 'model7/' + 'EVs/' + 'run{0}_fixed_before_cond_incorr.txt'.format(i * 2 + 1), 
                   run1_fixed_before_cond_incorr_mtrx, delimiter = '\t', fmt = '%.4f')     
        np.savetxt(sub_dir + '/MRthesis/' + 'model7/' + 'EVs/' + 'run{0}_fixed_same.txt'.format(i * 2 + 1), 
                   run1_fixed_same_mtrx, delimiter = '\t', fmt = '%.4f')        
        np.savetxt(sub_dir + '/MRthesis/' + 'model7/' + 'EVs/' + 'run{0}_fixed_change.txt'.format(i * 2 + 1), 
                   run1_fixed_change_mtrx, delimiter = '\t', fmt = '%.4f')       
        np.savetxt(sub_dir + '/MRthesis/' + 'model7/' + 'EVs/' + 'run{0}_fixed_lost.txt'.format(i * 2 + 1), 
                   run1_fixed_lost_mtrx, delimiter = '\t', fmt = '%.4f')           
        np.savetxt(sub_dir + '/MRthesis/' + 'model7/' + 'EVs/' + 'run{0}_nonresponse.txt'.format(i * 2 + 1), 
                   run1_nonresponse_mtrx, delimiter = '\t', fmt = '%.4f')           
        
   
        np.savetxt(sub_dir + '/MRthesis/' + 'model7/' + 'EVs/' + 'run{0}.txt'.format(i * 2 + 2), 
                   run2_mtrx, delimiter = '\t', fmt = '%.4f')
        np.savetxt(sub_dir + '/MRthesis/' + 'model7/' + 'EVs/' + 'run{0}_fixed_before_cond_corr.txt'.format(i * 2 + 2), 
                   run2_fixed_before_cond_corr_mtrx, delimiter = '\t', fmt = '%.4f')       
        np.savetxt(sub_dir + '/MRthesis/' + 'model7/' + 'EVs/' + 'run{0}_fixed_before_cond_incorr.txt'.format(i * 2 + 2), 
                   run2_fixed_before_cond_incorr_mtrx, delimiter = '\t', fmt = '%.4f')        
        np.savetxt(sub_dir + '/MRthesis/' + 'model7/' + 'EVs/' + 'run{0}_fixed_same.txt'.format(i * 2 + 2), 
                   run2_fixed_same_mtrx, delimiter = '\t', fmt = '%.4f')         
        np.savetxt(sub_dir + '/MRthesis/' + 'model7/' + 'EVs/' + 'run{0}_fixed_change.txt'.format(i * 2 + 2), 
                   run2_fixed_change_mtrx, delimiter = '\t', fmt = '%.4f')        
        np.savetxt(sub_dir + '/MRthesis/' + 'model7/' + 'EVs/' + 'run{0}_fixed_lost.txt'.format(i * 2 + 2), 
                   run2_fixed_lost_mtrx, delimiter = '\t', fmt = '%.4f')           
        np.savetxt(sub_dir + '/MRthesis/' + 'model7/' + 'EVs/' + 'run{0}_nonresponse.txt'.format(i * 2 + 2), 
                   run2_nonresponse_mtrx, delimiter = '\t', fmt = '%.4f')  
'''
```

# WMAZE Descriptives -- Double Fixed Trials

```python
from math import sqrt
%matplotlib inline
import seaborn as sns
sns.set_palette('muted')
import matplotlib.pyplot as plt

subs = ['WMAZE_001', 'WMAZE_002', 'WMAZE_003', 'WMAZE_004', 'WMAZE_005',
        'WMAZE_006', 'WMAZE_007', 'WMAZE_008', 'WMAZE_009', 'WMAZE_010', 
        'WMAZE_017', 'WMAZE_018', 'WMAZE_019', 'WMAZE_020', 'WMAZE_021',
        'WMAZE_022', 'WMAZE_023', 'WMAZE_024', 'WMAZE_026', 'WMAZE_027']

#subs = ['WMAZE_001']
# Array containing the three sets
stim_sets = ['set1', 'set2', 'set3']

count_table = {}
ctstd_table = {}
rt_table = {}
rtstd_table = {}

for sub in subs:
    ct_dict = {}
    rt_dict = {}
    sub_dir = '/home/data/madlab/data/mri/wmaze/scanner_behav/{0}/'.format(sub)
    dir_file = glob(join(sub_dir, '{0}_wmazebl_2015*.txt'.format(sub)))    
    dir_file.sort() 

    for i, curr_set in enumerate(stim_sets):

        run1 = pd.read_table(dir_file[i * 2])
        run2 = pd.read_table(dir_file[i * 2 + 1])
        
        run1 = run1[:-3]
        run2 = run2[:-3]
        
        # Convert dataframes into numpy arrays for trial type and correct/incorrect
        # With BLs
        run1_trialtype = run1['TrialType'].values
        run1_correct = run1['Correct'].values
        run1_resp = run1['Resp'].values 
        run2_trialtype = run2['TrialType'].values
        run2_correct = run2['Correct'].values
        run2_resp = run2['Resp'].values
        

        # Shift the TrialType column up by 1 and insert dummy element into the first index
        # Run 1
        run1_trial_shift = run1_trialtype[1:] 
        run1_trial_shift = np.append(run1_trial_shift, -1)
        # Run 2
        run2_trial_shift = run2_trialtype[1:]
        run2_trial_shift = np.append(run2_trial_shift, -1)

        # Shift the Correct column up by 1 and insert dummy element into the first index
        # Run 1
        run1_correct_shift = run1_correct[1:]
        run1_correct_shift = np.append(run1_correct_shift, -1)
        # Run 2
        run2_correct_shift = run2_correct[1:]
        run2_correct_shift = np.append(run2_correct_shift, -1)
       
        
        
        run1_2A_corr = np.where((run1_trial_shift == 'A') & (run1_trialtype == 'A') & (run1_correct_shift == 1))
        run1_2C_corr = np.where((run1_trial_shift == 'C') & (run1_trialtype == 'C') & (run1_correct_shift == 1))
        run1_AC_corr = np.where((run1_trial_shift == 'C') & (run1_trialtype == 'A') & (run1_correct_shift == 1))
        run1_CA_corr = np.where((run1_trial_shift == 'A') & (run1_trialtype == 'C') & (run1_correct_shift == 1))
        run1_same_corr = sorted(np.append(run1_2A_corr, run1_2C_corr))
        run1_change_corr = sorted(np.append(run1_AC_corr, run1_CA_corr))
        
        run1_2A_incorr = np.where((run1_trial_shift == 'A') & (run1_trialtype == 'A') & (run1_correct_shift == 0))
        run1_2C_incorr = np.where((run1_trial_shift == 'C') & (run1_trialtype == 'C') & (run1_correct_shift == 0))
        run1_AC_incorr = np.where((run1_trial_shift == 'C') & (run1_trialtype == 'A') & (run1_correct_shift == 0))
        run1_CA_incorr = np.where((run1_trial_shift == 'A') & (run1_trialtype == 'C') & (run1_correct_shift == 0))
        run1_same_incorr = sorted(np.append(run1_2A_incorr, run1_2C_incorr))
        run1_change_incorr = sorted(np.append(run1_AC_incorr, run1_CA_incorr))

        
        run2_2A_corr = np.where((run2_trial_shift == 'A') & (run2_trialtype == 'A') & (run2_correct_shift == 1))
        run2_2C_corr = np.where((run2_trial_shift == 'C') & (run2_trialtype == 'C') & (run2_correct_shift == 1))
        run2_AC_corr = np.where((run2_trial_shift == 'C') & (run2_trialtype == 'A') & (run2_correct_shift == 1))
        run2_CA_corr = np.where((run2_trial_shift == 'A') & (run2_trialtype == 'C') & (run2_correct_shift == 1))
        run2_same_corr = sorted(np.append(run2_2A_corr, run2_2C_corr))
        run2_change_corr = sorted(np.append(run2_AC_corr, run2_CA_corr))
        
        run2_2A_incorr = np.where((run2_trial_shift == 'A') & (run2_trialtype == 'A') & (run2_correct_shift == 0))
        run2_2C_incorr = np.where((run2_trial_shift == 'C') & (run2_trialtype == 'C') & (run2_correct_shift == 0))
        run2_AC_incorr = np.where((run2_trial_shift == 'C') & (run2_trialtype == 'A') & (run2_correct_shift == 0))
        run2_CA_incorr = np.where((run2_trial_shift == 'A') & (run2_trialtype == 'C') & (run2_correct_shift == 0))
        run2_same_incorr = sorted(np.append(run2_2A_incorr, run2_2C_incorr))
        run2_change_incorr = sorted(np.append(run2_AC_incorr, run2_CA_incorr))
        
        
        run_types = {'run1_same_corr': run1_same_corr, 'run1_change_corr': run1_change_corr,
                     'run2_same_corr': run2_same_corr, 'run2_change_corr': run2_change_corr,
                     'run1_same_incorr': run1_same_incorr, 'run1_change_incorr': run1_change_incorr,
                     'run2_same_incorr': run2_same_incorr, 'run2_change_incorr': run2_change_incorr} 
        
        for key in run_types:
            if len(run_types[key]) > 0:
                if run_types[key][-1] == 156:
                    run_types[key] = run_types[key][:-1]                            
        #print len(run_types['run1_same_corr'])
        run1_RTs = run1['RT']
        run2_RTs = run2['RT']    
         
        run1_same_corr_RTs = run1_RTs.values[(run_types['run1_same_corr'],)]
        run1_change_corr_RTs = run1_RTs.values[(run_types['run1_change_corr'],)] 
        run1_same_incorr_RTs = run1_RTs.values[(run_types['run1_same_incorr'],)]
        run1_change_incorr_RTs = run1_RTs.values[(run_types['run1_change_incorr'],)]
        
        run2_same_corr_RTs = run2_RTs.values[(run_types['run2_same_corr'],)]
        run2_change_corr_RTs = run2_RTs.values[(run_types['run2_change_corr'],)]
        run2_same_incorr_RTs = run2_RTs.values[(run_types['run2_same_incorr'],)]
        run2_change_incorr_RTs = run2_RTs.values[(run_types['run2_change_incorr'],)]
    
               
        for curr_type in ['same', 'change']:
            for acc in ['corr', 'incorr']:
                curr_name = '{0}_{1}'.format(curr_type, acc)
                rt_name = '{0}_{1}_RTs'.format(curr_type, acc)
                if not curr_name in ct_dict:
                    ct_dict[curr_name] = []
                ct_dict[curr_name].append(len(eval('run1_' + curr_name)))
                ct_dict[curr_name].append(len(eval('run2_' + curr_name)))
                if not rt_name in rt_dict:
                    rt_dict[rt_name] = []
                rt_eval = eval('run1_' + rt_name)
                rt_notNaN = np.where(rt_eval >= 0)
                rt_notNaN = rt_eval[rt_notNaN[0]]           
                if rt_notNaN.shape[0] == 0:
                    rt_dict[rt_name].append(None)                       
                else:
                    rt_dict[rt_name].append(np.average(rt_notNaN))

                rt_eval = eval('run2_' + rt_name)
                rt_notNaN = np.where(rt_eval >= 0)
                rt_notNaN = rt_eval[rt_notNaN[0]]           
                if rt_notNaN.shape[0] == 0:
                    rt_dict[rt_name].append(None)                       
                else:
                    rt_dict[rt_name].append(np.average(rt_notNaN))                   
    
    for key in ct_dict:
        ct_dict[key] = np.sum(ct_dict[key])
        if not key in count_table:
            count_table[key] = []
        count_table[key].append(ct_dict[key])
        
        
    for key in rt_dict:
        rt_notNONE = np.where(np.array(rt_dict[key]) >= 0)
        rt_dict[key] = np.average(np.array(rt_dict[key])[rt_notNONE[0]])
        if not key in rt_table:
            rt_table[key] = []
        rt_table[key].append(rt_dict[key])

        
df = pd.DataFrame(count_table, index = subs) 
df2 = pd.DataFrame(rt_table, index = subs) 
```
